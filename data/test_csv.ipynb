```python
import sys
import pandas as pd
import csv
import json
import cPickle as pickle

#with open("nn_train_datav3_test.json","r") as f:
#    org = json.load(f)
#filenames = org['filenames']
filenames = pickle.load(open('bin/filenames_test.bin', "rb"))
preds = pickle.load(open('bin/head-64x64_da_preds.bin', "rb"))

whale_ids=pd.read_csv("data/sample_submission.csv")
output = pd.DataFrame(columns=whale_ids.columns.values)
#print type(whale_ids['Image']) # => pandas.Series
print len(whale_ids['Image'])
print output

# get the dict
reverse_map = pickle.load(open("bin/label_map_reverse.bin", "rb"))

# convert label ids to whale ids and constru ct pandas.DataFrame obj(similar to csv structure)
# sample_submission-row-base output (keeps the original order)
for idx, filename in enumerate(whale_ids['Image']):
    row_dict={}
    # set zeros first
    for whale_id in whale_ids.columns.values[1:]:
        row_dict[whale_id] = 0

    # preds and filenames should have the same index for a sample
    index = filenames.index(filename)
    label = preds[index].argmax()
    whale_id = reverse_map[label]
    row_dict[whale_id] = 1
    row_dict['Image'] = filename
    output.loc[idx] = row_dict

    if idx % 1000 == 0:
        print('%d, %d' % (idx, label))

print output
output.to_csv('test.csv')
```

```python
import sys
import pandas as pd
import csv
import json
import cPickle as pickle

#with open("nn_train_datav3_test.json","r") as f:
#    org = json.load(f)
#filenames = org['filenames']
filenames = pickle.load(open('bin/filenames_test.bin', "rb"))
preds = pickle.load(open('bin/head_64x64_da_preds.bin', "rb"))

whale_ids=pd.read_csv("data/sample_submission.csv")
print len(whale_ids.columns.values)

output = pd.DataFrame(columns=whale_ids.columns.values)
#print type(whale_ids['Image']) # => pandas.Series
print len(whale_ids['Image'])
print output

# get the dicts
label_map = pickle.load(open("bin/label_map.bin", "rb"))            # class_name => label(integer)
reverse_map = pickle.load(open("bin/label_map_reverse.bin", "rb"))  # label(integer) => class_name

# preds-base output + probability output (export with decimals)
for idx, pred in enumerate(preds):
    # Initialize a row: set probabilities
    row_dict = {}
    for whale_id in whale_ids.columns.values[1:]: # whale_ids.columns[0] = 'Image'
        label = label_map[whale_id]
        prob = pred[label]
        row_dict[whale_id] = prob

    row_dict['Image']=filenames[idx]
    output.loc[idx] = row_dict

    if idx % 1000 == 0:
        print('%d, %d' % (idx, label))

print output
print len(output.columns)
output.to_csv('test_prob.csv')
```
